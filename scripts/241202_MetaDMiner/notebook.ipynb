```python
import metadynminer
hillsfile = metadynminer.Hills(name="/home/jakub/phd/openmm-md/data/241010_FoldingUponBinding/output/ASYN-A/241128-MetaD/A-synuclein_alpha.hills")
```

```python
fes = metadynminer.Fes(hillsfile)
```

```python
fes.plot(png_name="fes.png")
```

```python
minima = metadynminer.Minima(fes)
print(minima.minima)
minima.plot(png_name="fes.png")
```

```python
fep = metadynminer.FEProfile(minima, hillsfile)
fep.plot()
```
